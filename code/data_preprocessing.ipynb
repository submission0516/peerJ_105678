## Data Cleaning

```python
# # Extract the first date from the 'PI' column and convert it to datetime
# df1['Time'] = df1['PI'].str.extract('(\d{2}\s\w+\s\d{4})')
# df1['Time'] = pd.to_datetime(df1['Time'], errors='coerce')

# Extract the last date from the 'PI' column and convert it to datetime
# Function to extract the last date from the 'PI' column
def extract_last_date(row):
    dates = [date.split()[-3:] for date in row.split(';')]
    dates = [date for date in dates if len(date) == 3]
    if dates:
        return max([datetime.strptime(' '.join(date), '%d %b %Y') for date in dates])
    else:
        return pd.NaT

# Create the 'Time' column by applying the function
df1['Time'] = df1['PI'].apply(extract_last_date)


# Remove prefix "NOVELTY -" from the "AB" column
df1['AB'] = df1['AB'].str.replace('NOVELTY - ', '')

# Split the "AU" and "DC" columns based on the ";" delimiter
df1['AU'] = df1['AU'].str.split(';')
df1['DC'] = df1['DC'].str.split(';')

# Strip whitespaces from each string in the "AU" and "DC" columns
df1['AU'] = df1['AU'].apply(lambda x: [item.strip() for item in x] if isinstance(x, list) else [])
df1['DC'] = df1['DC'].apply(lambda x: [item.strip() for item in x] if isinstance(x, list) else [])


# 'TI' and 'AB' are the column we want to clean
df1['TI'] = df1['TI'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)
df1['AB'] = df1['AB'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)

# Assuming df1 is your DataFrame
df1 = df1.dropna(subset=['AB'])

# Assuming df1 is your DataFrame
df1 = df1.sort_values(by='Time', ascending=True)

# Drop rows with duplicate values in the "TI" column
df1 = df1.drop_duplicates(subset='TI', keep='first')

# Keep only the required columns
df1_cleaned = df1[['TI', 'AU', 'AB', 'DC', 'Time']]

# Filter the DataFrame to include only rows with dates within the specified range
start_date = '2013-01-01'
end_date = '2022-12-31'

df1_cleaned = df1_cleaned[df1_cleaned['Time'].between(start_date, end_date)]

# Display the resulting DataFrame
df1_cleaned.head(5)
```

## Data Cleaning

```python
# Select the desired columns
df2 = df2[['Article Title', 'Authors', 'Abstract', 'Source Title', 'Publication Date', 'Publication Year']]

# Custom function to create the "Time" column
def create_time(row):
    if pd.notna(row['Publication Date']):
        publication_date = str(row['Publication Date']).strip().upper()
    else:
        publication_date = None

    if pd.notna(row['Publication Year']):
        publication_year = int(row['Publication Year'])
    else:
        publication_year = None

    # Define a dictionary mapping month abbreviations to month numbers
    month_mapping = {'JAN': 1, 'FEB': 2, 'MAR': 3, 'APR': 4, 'MAY': 5, 'JUN': 6,
                     'JUL': 7, 'AUG': 8, 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DEC': 12}

    # Function to parse the publication date and return a datetime object
    def parse_date(date_str):
        if date_str in month_mapping:
            month = month_mapping[date_str]
        else:
            month = 1  # Default to January if month is not specified

        return datetime(publication_year, month, 1)

    # Use the custom function to create the "Time" column
    if publication_date:
        time_value = parse_date(publication_date)
    else:
        time_value = None

    return time_value

# Apply the custom function to create the "Time" column
df2['Time'] = df2.apply(create_time, axis=1)

# Fill missing values with the previous non-null value
df2['Time'].fillna(method='ffill', inplace=True)

# Drop unnecessary columns
df2 = df2[['Article Title', 'Authors', 'Abstract', 'Source Title', 'Time']]

# Rename columns as specified
df2 = df2.rename(columns={'Article Title': 'TI', 'Authors': 'AU', 'Abstract': 'AB', 'Source Title': 'DC', 'Time': 'Time'})


df2.head(5)
```

```python
# Split the "AU" and "DC" columns based on the ";" delimiter
df2['AU'] = df2['AU'].str.split(';')


# 'TI' and 'AB' are the column we want to clean
df2['TI'] = df2['TI'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)
df2['AB'] = df2['AB'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)

# Assuming df2 is your DataFrame
df2 = df2.dropna(subset=['AB', 'Time', 'TI'])

# Assuming df2 is your DataFrame
df2 = df2.sort_values(by='Time', ascending=True)

# Drop rows with duplicate values in the "TI" column
df2 = df2.drop_duplicates(subset='TI', keep='first')

# Keep only the required columns
df2_cleaned = df2[['TI', 'AU', 'AB', 'DC', 'Time']]

# Filter the DataFrame to include only rows with dates within the specified range
start_date = '2013-01-01'
end_date = '2022-12-31'

df2_cleaned = df2_cleaned[df2_cleaned['Time'].between(start_date, end_date)]

# Display the resulting DataFrame
df2_cleaned.head(5)
```
