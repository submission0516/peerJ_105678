## Data Cleaning

```python
df1 = df
```

```python
# Extract the first date from the 'PI' column and convert it to datetime
df1['Time'] = df1['PI'].str.extract('(\d{2}\s\w+\s\d{4})')
df1['Time'] = pd.to_datetime(df1['Time'], errors='coerce')

# # Extract the last date from the 'PI' column and convert it to datetime
# # Function to extract the last date from the 'PI' column
# def extract_last_date(row):
#     dates = [date.split()[-3:] for date in row.split(';')]
#     dates = [date for date in dates if len(date) == 3]
#     if dates:
#         return max([datetime.strptime(' '.join(date), '%d %b %Y') for date in dates])
#     else:
#         return pd.NaT

# # Create the 'Time' column by applying the function
# df1['Time'] = df1['PI'].apply(extract_last_date)


# Remove prefix "NOVELTY -" from the "AB" column
df1['AB'] = df1['AB'].str.replace('NOVELTY - ', '')

# Split the "AU" and "DC" columns based on the ";" delimiter
df1['AU'] = df1['AU'].str.split(';')
df1['DC'] = df1['DC'].str.split(';')

# Strip whitespaces from each string in the "AU" and "DC" columns
df1['AU'] = df1['AU'].apply(lambda x: [item.strip() for item in x] if isinstance(x, list) else [])
df1['DC'] = df1['DC'].apply(lambda x: [item.strip() for item in x] if isinstance(x, list) else [])


# 'TI' and 'AB' are the column we want to clean
df1['TI'] = df1['TI'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)
df1['AB'] = df1['AB'].apply(lambda x: re.sub(r'\s+', ' ', str(x)) if isinstance(x, (str, bytes)) else x)

# Assuming df1 is your DataFrame
df1 = df1.dropna(subset=['AB'])

# Assuming df1 is your DataFrame
df1 = df1.sort_values(by='Time', ascending=True)

# Drop rows with duplicate values in the "TI" column
df1 = df1.drop_duplicates(subset='TI', keep='first')

# Keep only the required columns
df1_cleaned = df1[['TI', 'AU', 'AB', 'DC', 'Time']]

# Filter the DataFrame to include only rows with dates within the specified range
start_date = '2011-01-01'
end_date = '2020-12-31'

df1_cleaned = df1_cleaned[df1_cleaned['Time'].between(start_date, end_date)]

# Display the resulting DataFrame
df1_cleaned.head(5)
```

```python
df['length'] = df['AB'].apply(lambda x: len(str(x)))

min_length = df['length'].min()
max_length = df['length'].max()

print("Minimum length:", min_length)
print("Maximum length:", max_length)
```
